```python
#여러개의 선 그리기 :plot() 여러번 사용
t=np.arange(0.,5.,0.2) # 0부터 5 '전'까지 0.2 간격으로 숫자를 만드는 함수 
t
plt.figure(figsize=(10,6)) #그래프 전체 그림판 크기 지정
plt.title('라인플롯에서 여러개의 선 그리기')
plt.plot(t,t,'r--') #plt.plt(x축, y축, 스타일 문자열)
#그래프 1
plt.plot(t,0.5*t**2,'bs:')
#그래프 2
plt.plot(t,0.2*t**3,'g^-')
#그래프 3
plt.show()
```

```python
plt.plot(t,t,'r--',t,0.5*t**2,'bs:',t,0.2*t**3,'g^-')
plt.show()
```

```python
#여러개의 선 그리기 
t=np.arange(0.,5.,0.2) # 0부터 5 '전'까지 0.2 간격으로 숫자를 만드는 함수 
t
plt.figure(figsize=(10,6)) #그래프 전체 그림판 크기 지정
plt.title('라인플롯에서 여러개의 선 그리기: plot 한 번 사용')

plt.plot(t,t,'r--',t,0.5*t**2,'bs:',t,0.2*t**3,'g^-')
plt.show()
```

```python
#2x2 형태의 네 갯의 플롯

np.random.seed(0) #랜덤 숫자 생성 순서 고정 (매번 코드 실행해도 같은 랜덤 값이 나옴)

plt.subplot(221)  #위치는 함수 전에 생성되어야 함
#2행 2열로 그림판을 나눈 후, 1번째 위치 
plt.plot(np.random.rand(5))
#랜덤 값 5개로 선 그래프
plt.title('axes1')
#서브플롯 제목

plt.subplot(222)  #위치는 함수 전에 생성되어야 함
plt.plot(np.random.rand(5))
plt.title('axes2')


plt.subplot(223)  #위치는 함수 전에 생성되어야 함
plt.plot(np.random.rand(5))
plt.title('axes3')


plt.subplot(224)  #위치는 함수 전에 생성되어야 함
plt.plot(np.random.rand(5))
plt.title('axes4')

plt.tight_layout(pad=1.5)
#서브플롯 간 간격 자동 조정, 그래프와 제목 사이 여백

#subplot(행,열,번호)
#plot() : 그래프 그림
#title(): 서브플롯 제목
#tight_layout() : 여백 자동 조정

plt.show()
```

```python
fig, ax = plt.subplots(2,2)
#plt.subplots(): 한번에 여러 서브 플롯만들기 
#ax[i,j]: 각 서브플롯(=작은 그래프_ 위치 지정

ax[0,0].plot(np.random.rand(5)) #왼쪽위 그래프, 랜덤 값 5개 선 그래프 
ax[0,0].set_title('axes1') #서브플롯 제목 설정

ax[0,1].plot(np.random.rand(5))
ax[0,1].set_title('axes2')

ax[1,0].plot(np.random.rand(5))
ax[1,0].set_title('axes2')

ax[1,1].plot(np.random.rand(5))
ax[1,1].set_title('axes1')

plt.tight_layout(pad=1.5) #서브플롯 간 간격 자동 조정
plt.show()
```

```python
#sin값 생성
y=np.sin(t)
#np.sin() : 넘파이에서 제공하는 사인 함수 
```

```python
plt.figure(figsize=(10,6))
plt.plot(t,y)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.title('Example of sinewave')
plt.plot(t,y)
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.title('Example of sine/cosine wave')
plt.plot(t,y)
plt.plot(t,np.cos(t), c ='red')
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
np.random.seed(0)
people = ['몽룡','춘향','방자','향단']
y=np.arange(len(people))

performance = 3+10 *np.random.rand(len(people))
#0~1난수 4개 생성 -> 0~10으로 확대 -> 3더하기 -> 3~13범위 난수
print(y)
print(performance)

plt.title('Bar Chart')
plt.barh(y,performance,alpha=0.4) #y 좌표에 따라 막대 그림, 길이 = performance값
#plt.barh(y,performance) : 수평 막대그래프에서 막대 길이로 사용됨 
plt.yticks(y,people)
plt.xlabel('xlabel')
plt.ylabel('ylabel')

plt.show()
```
